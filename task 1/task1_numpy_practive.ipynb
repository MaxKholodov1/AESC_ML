```python
import numpy as np
def res1(m):
  diagonal=m.diagonal()
  n=diagonal[diagonal!=0]
  pr=np.prod(n)
  return pr
```

```python
# code here
def res(x,n):
  ans=1
  for i in range(n):
    if(x[i][i]!=0):
      ans*=x[i][i]
  return ans;
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой.  
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ – 5.

```python
# code here
def max_after_zerononvect(x):
    first = False
    mx=0
    for i in range(1, len(x)):
        if(first==False and x[i-1]==0):
            mx=x[i]
        else:
            if(x[i-1]==0 and x[i]>mx):
                mx=x[i]
    return mx
```

```python
import numpy as np

def max_after_zerovect(x):
    zero_indices = np.where(x[:-1] == 0)[0]
    elements_after_zero = x[zero_indices + 1]
    return elements_after_zero.max() if elements_after_zero.size > 0 else 0
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np
time_vectorized=[]
time_non_vectorized=[]
s=[]
for i in range (1000, 100001, 100):
    s.append(i)
    x = np.random.randint(-50, 50, size=i)
    startnonvect=time.time()
    max_after_zerononvect(x)
    endnonvect= time.time()
    ansnonvect=endnonvect-startnonvect
    startvect=time.time()
    max_after_zerovect(x)
    endvect= time.time()
    ansvect=endvect-startvect
    time_vectorized.append(ansvect)
    time_non_vectorized.append(ansnonvect)
f, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(s, time_non_vectorized, label=u"Невекторизованная", marker='o')
ax2.plot(s, time_vectorized, label=u"Векторизованная", marker='o')
ax2.set_title(u"Сравнение времени выполнения")
ax2.set_xlabel(u"Размер массива")
ax2.set_ylabel(u"Время выполнения (с)")
ax2.grid()
ax2.legend()

f.show()
```

* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# code here
def run_length_encodingnon(x):
  
    values =[x[0]]
    counts=[1]
    for i in range(1, len(x)):
        if(x[i]==values[-1]):
            counts[-1]+=1
        else:
            values.append(x[i])
            counts.append(1)
    
    return values, counts

values, counts = run_length_encodingnon(x)
```

```python
import numpy as np

def run_length_encodingvec(x):
    changes = np.where(np.diff(x) != 0)[0] + 1
    values = np.concatenate(([x[0]], x[changes]))
    counts = np.diff(np.concatenate(([0], changes, [len(x)])))
    return values, counts
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np
time_vectorized=[]
time_non_vectorized=[]
s=[]
for i in range (100000, 500000, 10000):
    s.append(i)
    x = np.random.randint(-50, 50, size=i)
    startnonvect=time.time()
    run_length_encodingnon(x)
    endnonvect= time.time()
    ansnonvect=endnonvect-startnonvect
    startvect=time.time()
    run_length_encodingvec(x)
    endvect= time.time()
    ansvect=endvect-startvect
    time_vectorized.append(ansvect)
    time_non_vectorized.append(ansnonvect)
f, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(s, time_non_vectorized, label=u"Невекторизованная", marker='o')
ax2.plot(s, time_vectorized, label=u"Векторизованная", marker='o')
ax2.set_title(u"Сравнение времени выполнения")
ax2.set_xlabel(u"Размер массива")
ax2.set_ylabel(u"Время выполнения (с)")
ax2.grid()
ax2.legend()

f.show()
```

```python
import numpy as np
def distnumpy(X, Y):
    return np.linalg.norm(X[:, np.newaxis, :] - Y[np.newaxis, :, :], axis=2)
```

```python
import scipy
def euclidean_distance_scipy(X, Y):
    return scipy.spatial.distance.cdist(X, Y)
#sd
```

```python
import numpy as np
def distpy(X, Y):
    result = []
    for x in X:
        row = []
        for y in Y:
            dist = sum((xi - yi) ** 2 for xi, yi in zip(x, y)) ** 0.5
            row.append(dist)
        result.append(row)
    return result
```

```python
import matplotlib.pyplot as plt
import time
import numpy as np

time_vectorized = []
time_non_vectorized = []
scipyy = []
s = []

for i in range(1000, 2001, 100):
    np.random.seed(42)
    s.append(i)
    X = np.random.rand(i, 2)
    Y = np.random.rand(i, 2)
    
    startnonvect = time.time()
    distpy(X, Y)
    endnonvect = time.time()
    ansnonvect = endnonvect - startnonvect
    
    startvect = time.time()
    distnumpy(X, Y)
    endvect = time.time()
    ansvect = endvect - startvect

    startscipy = time.time()
    euclidean_distance_scipy(X, Y)
    endscipy= time.time()
    ansscipy = endscipy - startscipy
    

    
    time_vectorized.append(ansvect)
    time_non_vectorized.append(ansnonvect)
    scipyy.append(ansscipy)
    print(i)

f, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(s, time_non_vectorized, label=u"Невекторизованная", marker='o')
ax2.plot(s, time_vectorized, label=u"Векторизованная", marker='o')
ax2.plot(s, scipyy, label=u"scipy", marker='o')
ax2.set_title(u"Сравнение времени выполнения")
ax2.set_xlabel(u"Размер массива")
ax2.set_ylabel(u"Время выполнения (с)")
ax2.grid()
ax2.legend()

# Показать график
plt.show()
```
